# Plot Summary Statistics

```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_delays = pd.read_csv('../assets/df_punctuality_by_Carrier_Name.csv', sep="\t")
```

```python
df_delays = df_delays[df_delays['Carrier_Name'] != "EasyJet"]
```

```python
df_delays
```

```python
airline_colors = {
    'KLM Royal Dutch Airlines': u'royalblue',
    'Air Europa': u'palegreen',
    'Iberia': u'red',
    'Ryanair': u'yellow',
    'SAS': u'mediumblue',
    'Aer Lingus': u'darkgreen',
    'British Airways': u'dodgerblue',
    'Lufthansa': u'navy',
    'Norwegian': u'tomato',
    'Ethiopian': u'goldenrod',
    'LOT': u'cyan',
    'Vueling': u'orange',
    'Transavia': u'forestgreen'}
```

```python
#plt.figure()
```

```python
def plot_results(df_delays, airline_colors):
    plt.rcdefaults()
    cols = ['Delayed pct', 'median Delay', 'mean Delay', 'Occurrences']
    colors = list(map(lambda x: airline_colors.get(x, 'black'), list(df_delays['Carrier_Name'])))
    df_delays['Colors'] = colors
    y_pos = np.arange(len(df_delays))
    for _ in cols:
        _temp = df_delays.sort_values(_, ascending=False)
        fig, ax = plt.subplots()
        ax.barh(y_pos, _temp[_], align='center', color=_temp['Colors'])
        ax.set_yticks(y_pos)
        ax.set_yticklabels(_temp['Carrier_Name'].unique())
        ax.invert_yaxis()
        ax.set_xlabel(_)
```

```python
plot_results(df_delays, airline_colors)
```
